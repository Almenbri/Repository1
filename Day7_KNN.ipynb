```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_csv('./gdrive/My Drive/Mawhiba/student-mat.csv', sep =';')
df.head()
```

```python
df.shape
```

```python
dfn = df[['Medu', 'failures','goout','freetime','absences','studytime','G3']]
dfn.tail()
```

```python
#convert grade(0-20) to 0 (Fail) if grade < 10 and 1 (Pass) if grade >= 10
dfn['Grade'] = pd.cut(df.G3,bins=[-1,9,20],labels=[0,1])
dfn.tail()
```

```python
dfn.Grade.value_counts()
```

```python
from sklearn.neighbors import KNeighborsClassifier

X = dfn[['Medu', 'studytime', 'absences', \
        'freetime', 'goout', 'failures']]
y = dfn['Grade']

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X, y)

Medu = 4
studytime = 2
absences = freetime = goout = 1
failures = 0

val_prediction = knn.predict(np.array([[Medu, studytime, \
                                        absences, freetime,\
                                        goout, failures]]))

if(val_prediction):
  print("Prediction Result: The Student Will Pass.")
else:
  print("Prediction Result: The Student Will Fail.")
```

```python
from sklearn.naive_bayes import GaussianNB

X = dfn[['Medu', 'studytime', 'absences', \
        'freetime', 'goout', 'failures']]
y = dfn['Grade']

NB_model = GaussianNB()
NB_model.fit(X, y)

Medu = 4
studytime = 2
absences = freetime = goout = 1
failures = 0

val_prediction = NB_model.predict(np.array([[Medu, studytime, \
                                        absences, freetime,\
                                        goout, failures]]))

if(val_prediction):
  print("Prediction Result: The Student Will Pass.")
else:
  print("Prediction Result: The Student Will Fail.")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score, accuracy_score, precision_score

X = dfn[['Medu', 'failures','goout','freetime','absences','studytime']]
y = dfn['Grade']

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_X, train_y)
val_prediction = knn.predict(val_X)

print('Accuracy = ', int(round(accuracy_score(val_y, val_prediction)*100,0)),'%')
print('Precision = ', int(round(precision_score(val_y, val_prediction)*100,0)),'%')
print('Recall = ', int(round(recall_score(val_y, val_prediction)*100,0)),'%')
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score,accuracy_score, precision_score

X = dfn[['Medu', 'failures','goout','absences']]
y = dfn['Grade']

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)

NB_model = GaussianNB()
NB_model.fit(train_X, train_y)
val_prediction = NB_model.predict(val_X)
print('Accuracy = ', int(round(accuracy_score(val_y, val_prediction)*100,0)),'%')
print('Precision = ', int(round(precision_score(val_y, val_prediction)*100,0)),'%')
print('Recall = ', int(round(recall_score(val_y, val_prediction)*100,0)),'%')
```
